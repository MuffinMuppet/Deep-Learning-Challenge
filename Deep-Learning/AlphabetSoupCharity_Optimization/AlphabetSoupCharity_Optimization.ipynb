```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
num_unique_value = application_df.nunique()
print(num_unique_value)
```

```python
# Look at APPLICATION_TYPE value counts for binning

application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

cutoff= 100  

application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(application_type_counts[application_type_counts < cutoff].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_over_one = classification_counts[classification_counts > 1]
print(classification_over_one)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 100
classification_counts = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = list(classification_counts[classification_counts < cutoff_value].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Convert categorical data to numeric
application_df_numeric = pd.get_dummies(application_df)

# Display the encoded dataframe
application_df_numeric.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_numeric.drop('IS_SUCCESSFUL', axis=1)
y = application_df_numeric['IS_SUCCESSFUL']
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

**## Optimizing the Model**


1.   Increased epochs to 150 and then 200.
2.   Then add neurons to a hidden layer.
3.   Add more hidden layers.





```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

no_input_features = X_train_scaled.shape[1]

nn_prime = tf.keras.models.Sequential()

# First hidden layer
nn_prime.add(tf.keras.layers.Dense(units=90, activation="relu", input_dim= no_input_features))

# Second hidden layer
nn_prime.add(tf.keras.layers.Dense(units=50, activation="relu"))

# Third hidden layer
nn_prime.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn_prime.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_prime.summary()
```

```python
# Compile the model
nn_prime.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn_prime.fit(X_train_scaled, y_train, epochs=150)
```

```python
model_loss, model_accuracy = nn_prime.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Train the model
fit_model = nn_prime.fit(X_train_scaled, y_train, epochs=200)
```

```python
model_loss, model_accuracy = nn_prime.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

no_input_features = X_train_scaled.shape[1]

nn_prime2 = tf.keras.models.Sequential()

#  Added more layers

# First hidden layer
nn_prime2.add(tf.keras.layers.Dense(units=128, activation="relu", input_dim= no_input_features))

# Second hidden layer
nn_prime2.add(tf.keras.layers.Dense(units=64, activation="relu"))

# Third hidden layer

nn_prime2.add(tf.keras.layers.Dense(units=32, activation="relu"))


# Output layer
nn_prime2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_prime2.summary()
```

```python
# Compile the model
nn_prime2.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model = nn_prime2.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn_prime2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
no_input_features = X_train_scaled.shape[1]

nn_prime3 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=256, activation='relu', input_dim=no_input_features),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model
nn_prime3.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
fit_model = nn_prime3.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn_prime3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
